```python
#checking the average  length of name
def column_name_lenghts(dataframe_columns):
    import re
    """Here we are manipulating the the given values such that if the overall data set has very long name then we are converting.
    It into short name by splitting them with the help of regex module
    """
    length_list=[len(val) for val in dataframe_columns]
    long_name=[val   for val in dataframe_columns if len(val)>round(np.mean(length_list)) and np.mean(length_list)>np.median(length_list) ]
    print(f"""
    Mean length {np.mean(length_list)} character
    Max length {max(length_list)} character
    Median length {np.median(length_list)} character
    Total number of values with long name: {len(long_name)} 
    """) 
    short_name=[re.split("[(:]", item)[0].strip() for item in long_name]
    replace_list=[]
    return long_name,short_name

#Substituting the long names to its respective sectors generic name 
lis,lis1=column_name_lenghts(df['Sector'])
```

```python
#copying the data set to new variable
df_new=df.copy()
```

```python
#Shorting long data name
lis,lis1=column_name_lenghts(df_new['Sector'])
df_new['Sector']=df_new['Sector'].replace(lis,lis1)
df_new
```

```python
#converting pivot data into tabular form
df_melt =  df_new.melt(id_vars="Sector",value_name="FDI",var_name="Year")
```

##Data collection

```python
#As we don't have the exachanges rates for this investments we need to get those exchange rates from officai website of reserve bank of india 
# website url https://m.rbi.org.in/scripts/ReferenceRateArchive.aspx
df_1=pd.read_csv("/content/drive/MyDrive/Financial_analysis/BankWise.csv")
df_1['Date']=pd.to_datetime(df_1['Date'])
df_1['Date']=df_1['Date'].dt.year
df_1=pd.DataFrame(df_1.groupby('Date').mean().reset_index())
df_1
```

```python
df_melt['Exchange_rate']="0"
for year in df_1['Date']:
  for ind,yrs in enumerate(df_melt['Year']):
    if yrs.split("-")[0]==str(year):  #here we are matching and joining the exchange rate for currency
      df_melt['Exchange_rate'].iloc[ind]=float(df_1[df_1['Date']==year]['USD'])
    else:
      pass
df_melt
```

```python
df_melt.rename(columns={"FDI":"FDI(in USD)"},inplace=True)
df_melt['FDI(in INR)']=df_melt['FDI(in USD)']*df_melt['Exchange_rate']
df_melt['FDI(in INR)']=df_melt['FDI(in INR)'].astype(float)
df_melt['Exchange_rate']=df_melt['Exchange_rate'].astype(float)
```

```python
df_melt.to_csv("Transformed_FDI_india.csv")
```

#Year wise trend analysis

##Data visualization

```python
fig,axes=plt.subplots(1,3)
df_melt.groupby('Year').sum().plot(kind="bar",y="FDI(in INR)",ax=axes[0],figsize=(30,5))
df_melt.groupby('Year').mean().plot(kind="line",y="Exchange_rate",ax=axes[1],figsize=(30,5))
df_melt.groupby('Year').sum().plot(kind="bar",y="FDI(in USD)",ax=axes[2],figsize=(30,5))
plt.show()
```

```python
sector_wise_group =df_melt.groupby(['Sector','Year']).sum().reset_index()
for i in sector_wise_group['Year'].unique():
  sector_wise_group[sector_wise_group["Year"]==i].plot(kind="bar",x="Sector",y="FDI(in INR)",figsize=(20,5),title=i)
  plt.show()
```

```python
list_of_year=[]
list_of_maximum_FDI=[]
for sec in df_melt['Sector'].unique():
  list_of_year.append(int(list(df_melt[df_melt['FDI(in INR)']==(df_melt[df_melt['Sector']==sec]['FDI(in INR)'].max())]['Year'])[0].split('-')[0])+1)
  list_of_maximum_FDI.append(round(df_melt[df_melt['Sector']==sec]['FDI(in INR)'].max()))
MAX_data=pd.DataFrame(dict(zip(['Sector','Year','FDI'],[list(df_melt['Sector'].unique()),list_of_year,list_of_maximum_FDI])))


MAX_data.groupby('Year').sum().reset_index().plot(kind="bar",x='Year',y='FDI',title='Maximum investments in Year',figsize=(25,10))
```
